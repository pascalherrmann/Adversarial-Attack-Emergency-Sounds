```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
```

```python
result= {'Animal': 0.006292650918635171,
 'Motor vehicle (road)': 0.007053805774278216,
 'Music': 0.006148293963254593,
 'Natural sounds': 0.005354330708661418,
 'Human voice': 0.004435695538057743}
result = {re.sub(r" ?\([^)]+\)", "", entry): [result[entry]*100] for entry in sorted(result, key = lambda x: result[x], reverse=True)}
```

```python
df = pd.DataFrame.from_dict(result)

sns.set()
sns.set_style("whitegrid")
sns.set_context("poster")

f = plt.figure(figsize=(15,10))
plt.ylabel("average success rate")
ax = sns.barplot(data=df)
```

```python
import numpy as np

sns.set()
sns.set_style("whitegrid")
sns.set_context("poster")

def sinplot(freq, inner, outer, color):
    x = np.linspace(0, freq, 100)
    plt.plot(x, outer*np.sin(inner*x), color=color)
```

```python
with sns.axes_style('white'):
    x = np.linspace(0, 30, 100)
    plt.plot(x, 2*np.sin(x/3) + -np.sin(x/2), color='m')
```

```python
with sns.axes_style('white'):
    x = np.linspace(0, 30, 100)
    plt.plot(x, -np.sin(x/2), color='r')
```

```python
with sns.axes_style('white'):
    x = np.linspace(0, 30, 100)
    plt.plot(x, 2*np.sin(x/3), color='b')
```
